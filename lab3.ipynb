```python
from functools import update_wrapper
import time
```

```python
class DecoratingTimer:
  def __init__(self, func):
    print('Dekoracja')
    self.func = func
    self.call_counter = 0
    self.czasy = []
    update_wrapper(self, func) # zachowuje metadane funkcji

  def __call__(self, *args, **kwargs):
    start = time.perf_counter()
    self.call_counter += 1
    print(f'Przed. Licznik: {self.call_counter}')
    res = self.func(*args, **kwargs)
    print(f'Po. Licznik: {self.call_counter}')
    end = time.perf_counter()
    t = end - start
    print("Czas trwania = %s s" % (t))
    self.czasy.append(t)
    return res

  def stats(self):
    n = self.func.__name__

    w = self.call_counter

    k = 0

    d = 0

    for i in range(self.call_counter):
      if self.czasy[i] < k or k == 0: k = self.czasy[i]
      if self.czasy[i] > d: d = self.czasy[i]

    s = sum(self.czasy) / self.call_counter

    return {'nazwa':n, 'liczba wywołań':w, 'najkrótszy czas':k, 'najdłuższy czas':d, 'średni czas':s}
```

```python
@DecoratingTimer
def funkcja(a, b):
  """Wyświetla liczby w odwróconej kolejności i zwraca ich iloczyn"""
  print(b, a)
  return a * b

funkcja(1, 2)

funkcja(1, 2)

funkcja(1, 2)
```

```python
funkcja.call_counter, funkcja.czasy, funkcja.stats()
```

```python
funkcja.__name__, funkcja.__doc__
```

```python
@DecoratingTimer
def wait(a):
  """Czeka podaną ilość sekund i wyświetla 'zZz'"""
  print('zZz '*a)
  time.sleep(a)

wait(1)

wait(2)

wait(3)
```

```python
wait.call_counter, wait.czasy, wait.stats()
```

```python
wait.__name__, wait.__doc__
```

```python
class Temperature:
  def __init__(self, celsius):
    self.celsius = celsius

  def __str__(self):
      return (f"{self._celsius}\u00B0C")

  @property
  def celsius(self):
    #print('Get C')
    return self._celsius

  @celsius.setter
  def celsius(self, celsius):
    #print('Set C')
    if celsius > -273.15: self._celsius = celsius
    else: self._celsius = -273.15

  @property
  def kelvin(self):
    #print('Get K')
    return round(self._celsius + 273.15, 2)

  @kelvin.setter
  def kelvin(self, kelvin):
    #print('Set K')
    if kelvin > 0: self._celsius = round(kelvin - 273.15, 2)
    else: self._celsius = -273.15

  @property
  def fahrenheit(self):
    #print('Get F')
    return round(9 * self._celsius / 5 + 32, 2)

  @fahrenheit.setter
  def fahrenheit(self, fahrenheit):
    #print('Set F')
    if fahrenheit > -459.67: self._celsius = round(5 * (fahrenheit - 32) / 9, 2)
    else: self._celsius = -273.15

  @property
  def freezing(self):
    return self._celsius <= 0

  @property
  def boiling(self):
    return self._celsius >= 100

  @property
  def state(self):
    if self._celsius < 0: return 'solid'
    if self._celsius < 100: return 'liquid'
    return 'gas'
```

```python
t = Temperature(25)

print(t.celsius)      # 25
print(t.fahrenheit)   # 77.0
print(t.kelvin)       # 298.15
print(t.state)
print(t.freezing)
print(t.boiling)
print()

t.fahrenheit = 32
print(t.celsius)      # 0.0
print(t.state)
print(t.freezing)
print(t.boiling)
print()

t.kelvin = 400
print(t.celsius)      # 126.85
print(t.state)
print(t.freezing)
print(t.boiling)
print()

t.kelvin = -300
print(t.kelvin)      # 0
print(t.state)
print(t.freezing)
print(t.boiling)
print()

print(t)
```

```python
def rng(seed, a, c, m, N, tryb='norm'):
  if tryb == 'norm':
    for i in range(N):
      yield seed
      seed = (a * seed + c) % m

  if tryb == '<1':
    for i in range(N):
      yield seed / m
      seed = (a * seed + c) % m
```

```python
for i in rng(1, 5, 3, 16, 5):
  print(i)
```

```python
gen = rng(1, 5, 3, 16, 5)
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
for i in rng(1, 5, 3, 16, 5, tryb='<1'):
  print(i)
```
